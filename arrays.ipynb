```python
keys = []
values = []
for i in dict:
    keys.append(i)
    values.append(dict[i])

with open(r'targets.txt', 'w') as fp:
    for item in keys:    
        fp.write(f"{item} \n")

with open(r'features.txt', 'w') as fp:
    for item in values:    
        fp.write(f"{item} \n")
```

```python
import matplotlib.pyplot as plt

features_array = np.loadtxt('features.txt')
targets_array = np.loadtxt('targets.txt')

# fig = plt.figure(figsize=(10,10))
fig, axs = plt.subplots(5,2, figsize = (15,10))
# axs = plt.subplots(5,2)

axs[0,0].scatter(features_array[:,0], targets_array)
axs[0,0].set_title('Upper bounds, points 1-5')
axs[0,1].scatter(features_array[:,1], targets_array)
axs[0,1].set_title('Lower bounds, points 1-5')
axs[1,0].scatter(features_array[:,2], targets_array)
axs[1,1].scatter(features_array[:,3], targets_array)
axs[2,0].scatter(features_array[:,4], targets_array)
axs[2,0].set_ylabel('Maximum x Velocity')
axs[2,1].scatter(features_array[:,5], targets_array)
axs[3,0].scatter(features_array[:,6], targets_array)
axs[3,1].scatter(features_array[:,7], targets_array)
axs[4,0].scatter(features_array[:,8], targets_array)
axs[4,1].scatter(features_array[:,9], targets_array)

for ax in axs.flat:
    ax.label_outer()
```
